# *Baseline Model 1*: ElasticNet

### In order to establish a baseline understanding of predicting drug sensitivity, I first explore prediction through ElasticNet. ElasticNet is ideal because it combines both L1 (LASSO) and L2 (Ridge) penalties for regularization, ensuring only the most important features are chosen while also assisting in overfitting concerns. ElasticNet is a linear model, making it easy to interpret.

```python
import sys
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import pickle
import os
sys.path.append("../src")

from plotting import plot_pred_vs_obs
from elasticnet import run_elasticnet
from feature import prepare_features, get_drug_data

df = pd.read_parquet("../data/df_clean.parquet")
mut_df = pd.read_parquet("../data/mut_df_clean.parquet")
```

```python
drug_features = {}
drugs = ["Cisplatin", "PLX-4720", "Selumetinib", "Trametinib", "Nutlin-3a (-)"]

for drug in drugs:
    X, X_mut, y = get_drug_data(df, mut_df, drug)
    drug_features[drug] = {"X": X, "X_mut": X_mut, "y": y}
```

## *Linear Baseline Model:* ElasticNet

```python
os.makedirs("saved_results", exist_ok=True)

drug_results = {}

for drug in drugs:
    X = drug_features[drug]["X"]
    y = drug_features[drug]["y"]
    
    print("="*60)
    print(f"Drug Name: {drug}")
    df_results = run_elasticnet(X, y, max_samples=None)
    
    drug_results[drug] = df_results
    with open(f"saved_results/{drug}_enet.pkl", "wb") as f:
        pickle.dump(df_results, f)
    print("="*60)
```

```python
os.makedirs("saved_results", exist_ok=True)
mut_drug_results = {}

for drug in drugs:
    X_mut = drug_features[drug]["X_mut"]
    y = drug_features[drug]["y"]
    
    print("="*60)
    print(f"Drug Name: {drug}")
    mut_df_results = run_elasticnet(X_mut, y, max_samples=None)   
    mut_drug_results[drug] = mut_df_results
    with open(f"saved_results/{drug}_mut_enet.pkl", "wb") as f:
        pickle.dump(mut_df_results, f)
    print("="*60)
```

```python
# Rerunning Trametinib separately to converge but not rewrite progress of the other drugs
# Increased max_iter from 2000 to 50000

os.makedirs("saved_results", exist_ok=True)

drugs = ["Trametinib"]

tram_drug_results = {}

for drug in drugs:
    X = drug_features[drug]["X"]
    y = drug_features[drug]["y"]
    
    print("="*60)
    print(f"Drug Name: {drug}")
    tram_df_results = run_elasticnet(X, y, max_samples=None)
    
    tram_drug_results[drug] = tram_df_results
    with open(f"saved_results/{drug}_enet.pkl", "wb") as f:
        pickle.dump(tram_df_results, f)
    print("="*60)
    
    
    
os.makedirs("saved_results", exist_ok=True)
tram_mut_drug_results = {}

for drug in drugs:
    X_mut = drug_features[drug]["X_mut"]
    y = drug_features[drug]["y"]
    
    print("="*60)
    print(f"Drug Name: {drug}")
    tram_mut_df_results = run_elasticnet(X_mut, y, max_samples=None)   
    tram_mut_drug_results[drug] = tram_mut_df_results
    with open(f"saved_results/{drug}_mut_enet.pkl", "wb") as f:
        pickle.dump(tram_mut_df_results, f)
    print("="*60)
```

```python
import pickle
import numpy as np

# Load all results
files = {
    "Cisplatin": ("saved_results/Cisplatin_enet.pkl", "saved_results/Cisplatin_mut_enet.pkl"),
    "PLX-4720": ("saved_results/PLX-4720_enet.pkl", "saved_results/PLX-4720_mut_enet.pkl"),
    "Selumetinib": ("saved_results/Selumetinib_enet.pkl", "saved_results/Selumetinib_mut_enet.pkl"),
    "Trametinib": ("saved_results/Trametinib_enet.pkl", "saved_results/Trametinib_mut_enet.pkl"),
    "Nutlin-3a (-)": ("saved_results/Nutlin-3a (-)_enet.pkl", "saved_results/Nutlin-3a (-)_mut_enet.pkl")
}

for drug, (f_standard, f_mut) in files.items():
    with open(f_standard, "rb") as f:
        standard = pickle.load(f)
    with open(f_mut, "rb") as f:
        mut = pickle.load(f)
    
    # Each .fit_time may be a list or array of times (e.g., from CV folds)
    standard_avg = np.mean(standard.get("fit_time", []))
    mut_avg = np.mean(mut.get("fit_time", []))
    
    print(f"{drug}:")
    print(f"  ElasticNet standard average fit time: {standard_avg:.4f} s")
    print(f"  ElasticNet + mutation average fit time: {mut_avg:.4f} s\n")
```

## Takeaways:

### Using ElasticNet, trametinib has had the best performance. Even as a baseline, a Spearman coefficient of > .7 indicates a very predictable drug and shows it outperformed a scientific article's results. This might lead us to believe that a linear model will be useful to implement in the final model.  Interestingly enough, trametinib needed many more iterations to converge compared to the other 4 drugs.

### Cisplatin is (as of now) an unpredictable drug, but maybe a more nonlinear approach will reveal new representations for this drug specifically.

### The ablation study shows that a combination of alpha = 0.05 and 0.1 and l1_ratio of 0.5 and 0.75 warrant the best model results.

### ElasticNet is very fast, with most fit times being under 1 second.
