# Creating an Optimal Model for Drug Sensitivity Prediction - With Mutation Data

### Since ElasticNet achieves respectable R2, Pearson, and Spearman metrics while remaining computationally efficient, I hypothesize that combining ElasticNet with a Multilayer Perceptron (MLP) will balance the benefits of linear modeling with the ability of the MLP to capture latent, potentially nonlinear patterns that a RandomForest might reveal.

## Mulilayer Perceptron

```python
import sys
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import pickle
import os
import torch
import torch.nn as nn
from torch.utils.data import TensorDataset, DataLoader, random_split
import torch.optim as optim
from torchvision.transforms import ToTensor
import time
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

sys.path.append("../src")
from plotting import plot_pred_vs_obs, analyze_drug_model_results
from elasticnet import run_elasticnet
from feature import prepare_features, get_drug_data
from mlp import get_device, plot_training_curves, train_model, test_model

df = pd.read_parquet("../data/df_clean.parquet")
mut_df = pd.read_parquet("../data/mut_df_clean.parquet")
```

```python
drug_features = {}
drug_loaders = {}
drugs = ["Cisplatin", "PLX-4720", "Selumetinib", "Trametinib", "Nutlin-3a (-)"]

for drug in drugs:
    print(f"Preparing data for: {drug}")
    X, X_mut, y = get_drug_data(df, mut_df, drug)
    drug_features[drug] = {"X": X, "X_mut": X_mut, "y": y}
    
    #X = drug_features[drug]["X"].values
    X_mut = drug_features[drug]["X_mut"].values
    y = drug_features[drug]["y"].values
    
    # ---- Train/Val vs Test split ----
    X_trainval, X_test, y_trainval, y_test = train_test_split(
        X_mut, y, test_size=0.1, random_state=2025
    )
    
    # ---- Scale ----
    scaler = StandardScaler()
    X_trainval_s = scaler.fit_transform(X_trainval)
    X_test_s = scaler.transform(X_test)
    
    # ---- Convert to tensors ----
    X_trainval_t = torch.tensor(X_trainval_s, dtype=torch.float32)
    y_trainval_t = torch.tensor(y_trainval, dtype=torch.float32).view(-1, 1)
    
    X_test_t = torch.tensor(X_test_s, dtype=torch.float32)
    y_test_t = torch.tensor(y_test, dtype=torch.float32).view(-1, 1)
    
    # ---- Full train dataset ----
    #train_data = TensorDataset(X_trainval_t, y_trainval_t)
    
    # ---- Split train into train/val ----
    #train_size = int(0.8 * len(train_data))
    train_size = int(0.8 * len(X_trainval_t))
    val_size = len(X_trainval_t) - train_size

    #train_set, val_set = random_split(train_data, [train_size, val_size],generator=torch.Generator().manual_seed(2025))
    
    train_set, val_set = random_split(
        TensorDataset(X_trainval_t, y_trainval_t),
        [train_size, val_size],
        generator=torch.Generator().manual_seed(2025)
    )
    
    # ---- Test dataset ----
    #test_data = TensorDataset(X_test_t, y_test_t)
    test_set = TensorDataset(X_test_t, y_test_t)
    
    # ---- DataLoaders ----
    train_loader = DataLoader(train_set, batch_size=32, shuffle=True, drop_last=False)
    val_loader   = DataLoader(val_set, batch_size=32, shuffle=False, drop_last=False)
    test_loader  = DataLoader(test_set, batch_size=32, shuffle=False, drop_last=False)
    
    # ---- Store everything ----
    drug_loaders[drug] = {
        "scaler": scaler,
        "train_loader": train_loader,
        "val_loader": val_loader,
        "test_loader": test_loader,
        "X_test": X_test_s,
        "X_test_t": X_test_t,
        "y_test": y_test
    }
```

```python
class MLP(nn.Module):
    def __init__(self, input_dim):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(input_dim, 128),
            nn.ReLU(),
            nn.Dropout(0.1),

            nn.Linear(128, 64),
            nn.ReLU(),
            nn.Dropout(0.1),

            nn.Linear(64, 1)
        )

    def forward(self, x):
        return self.net(x)
```

```python
drug_models = {}

device = get_device(use_gpu=True)
print("Device used:", device)

for drug in drugs:
    print(f"Training model for {drug}")
    loaders = drug_loaders[drug]
    input_dim = input_dim = drug_features[drug]["X_mut"].shape[1]
    
    model = MLP(input_dim=input_dim).to(device)
    loss_fn = nn.MSELoss()
    optimizer = optim.Adam(model.parameters(), lr=1e-4, weight_decay=1e-4)

    model, train_losses, val_losses, training_time = train_model(model, loaders["train_loader"], loaders["val_loader"], loss_fn, optimizer)
    
    drug_models[drug] = model

    print(f"Training losses for {drug} per epoch:", train_losses)
    print(f"Validation losses for {drug} per epoch:", val_losses)
    print(f"Total training time for {drug} (s):", training_time)
    print('-'*60)
```

```python
for drug, model in drug_models.items():
    test_loader = drug_loaders[drug]["test_loader"]
    
    metrics, all_preds, all_labels = test_model(model, test_loader)
    print(f"{drug} metrics:", metrics)
    print("-"*150)
    
    all_labels_flat = all_labels.flatten()
    all_preds_flat = all_preds.flatten()
    
    plot_pred_vs_obs(all_labels_flat, all_preds_flat, 
                     title=f"{drug} NN: Observed vs Predicted")
```

### There is variation , but overall the MLP seems to be performing as expected: typically better than both baseline models. Further research might reveal why Trametinib and Nutlin-3a (-) perform suboptimally within the MLP compared to the baseline models.

## Ensemble Model: Multilayer Perceptron & ElasticNet

```python
drugs = ["Cisplatin", "PLX-4720", "Selumetinib", "Trametinib", "Nutlin-3a (-)"]

# --- Initialize dictionaries ---
drug_mut_enets = {}

# --- Load all pipelines into dictionaries ---
for drug in drugs:
    # Mutation ENet
    with open(f"saved_results/{drug}_mut_enet.pkl", "rb") as f:
        drug_mut_enets[drug] = pickle.load(f)
```

```python
def ensemble_all_drugs(drugs, drug_enets, drug_models, drug_loaders):
    """
    Ensemble ElasticNet + NN preserving duplicate rows.
    Assumes NN scaler was fit on same columns as ElasticNet; aligns by order.
    """
    from sklearn.metrics import r2_score, mean_squared_error
    from scipy.stats import pearsonr, spearmanr
    import numpy as np
    import torch
    
    results = {}
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    
    for drug in drugs:
        print(f"\n{'='*60}")
        print(f"Processing: {drug}")
        print(f"{'='*60}")
        
        # --- Get ElasticNet data (with duplicates) ---
        enet_data = drug_enets[drug]
        pipeline = enet_data['pipeline']
        X_test_enet = enet_data['X_test']
        y_test_enet = enet_data['y_test']
        
        print(f"ElasticNet test set: {X_test_enet.shape}")
        
        # --- ENet prediction ---
        enet_pred = pipeline.predict(X_test_enet)
        print(f"ElasticNet predictions: {len(enet_pred)}")
        
        # --- Prepare NN test data ---
        nn_scaler = drug_loaders[drug]["scaler"]
        X_test_np = X_test_enet.values if hasattr(X_test_enet, 'values') else X_test_enet
        
        if drug == "Trametinib":
            n_features = 1164
            missing_cols = n_features - X_test_np.shape[1]
            if missing_cols > 0:
                X_test_np = np.hstack([X_test_np, np.zeros((X_test_np.shape[0], missing_cols))])
                print(f"Warning: Padded {missing_cols} missing features for {drug}")
            elif missing_cols < 0:
                X_test_np = X_test_np[:, :n_features]
                print(f"Warning: Trimmed {-missing_cols} extra features for {drug}")
        else:
            n_features = nn_scaler.scale_.shape[0]
            if X_test_np.shape[1] > n_features:
                X_test_np = X_test_np[:, :n_features]
                
            elif X_test_np.shape[1] < n_features:
                raise ValueError(f"NN scaler expects {n_features} features but X_test has {X_test_np.shape[1]}")
        
        X_test_nn_scaled = nn_scaler.transform(X_test_np)
        X_test_nn_tensor = torch.tensor(X_test_nn_scaled, dtype=torch.float32).to(device)
        
        # --- NN predictions ---
        nn_model = drug_models[drug]
        if isinstance(nn_model, dict):
            nn_model = nn_model["model"]
        
        nn_model.to(device)
        nn_model.eval()
        with torch.no_grad():
            nn_pred = nn_model(X_test_nn_tensor).cpu().numpy().flatten()
        
        print(f"NN predictions: {len(nn_pred)}")
        
        # --- Verify alignment ---
        assert len(enet_pred) == len(nn_pred) == len(y_test_enet), \
            f"Length mismatch! ElasticNet: {len(enet_pred)}, NN: {len(nn_pred)}, y_test: {len(y_test_enet)}"
        
        print(f" All predictions aligned: {len(enet_pred)} samples (including duplicates)")
        
        # --- Ensemble ---
        ensemble_pred = 0.5 * enet_pred + 0.5 * nn_pred
        y_test_array = y_test_enet.values if hasattr(y_test_enet, 'values') else y_test_enet
        
        # --- Metrics ---
        results[drug] = {
            "enet_r2": r2_score(y_test_array, enet_pred),
            "nn_r2": r2_score(y_test_array, nn_pred),
            "ensemble_r2": r2_score(y_test_array, ensemble_pred),
            "enet_rmse": np.sqrt(mean_squared_error(y_test_array, enet_pred)),
            "nn_rmse": np.sqrt(mean_squared_error(y_test_array, nn_pred)),
            "ensemble_rmse": np.sqrt(mean_squared_error(y_test_array, ensemble_pred)),
            "enet_pearson": pearsonr(y_test_array, enet_pred)[0],
            "nn_pearson": pearsonr(y_test_array, nn_pred)[0],
            "ensemble_pearson": pearsonr(y_test_array, ensemble_pred)[0],
            "enet_spearman": spearmanr(y_test_array, enet_pred)[0],
            "nn_spearman": spearmanr(y_test_array, nn_pred)[0],
            "ensemble_spearman": spearmanr(y_test_array, ensemble_pred)[0],
            "n_samples": len(y_test_array),
            "n_unique": len(np.unique(y_test_array)),
            "y_test": y_test_array,
            "y_pred_enet": enet_pred,
            "y_pred_nn": nn_pred,
            "y_pred_ensemble": ensemble_pred,
        }
        
        print(f"\nResults:")
        print(f"  ElasticNet R: {results[drug]['enet_r2']:.4f}, Pearson: {results[drug]['enet_pearson']:.4f}, Spearman: {results[drug]['enet_spearman']:.4f}")
        print(f"  NN R:         {results[drug]['nn_r2']:.4f}, Pearson: {results[drug]['nn_pearson']:.4f}, Spearman: {results[drug]['nn_spearman']:.4f}")
        print(f"  Ensemble R:   {results[drug]['ensemble_r2']:.4f}, Pearson: {results[drug]['ensemble_pearson']:.4f}, Spearman: {results[drug]['ensemble_spearman']:.4f}")
        improvement = results[drug]['ensemble_r2'] - max(results[drug]['enet_r2'], results[drug]['nn_r2'])
        print(f"  Improvement:   {improvement:+.4f}")
        print(f"  Samples:       {results[drug]['n_samples']} (including duplicates)")
    
    return results
```

```python
ensemble_results = ensemble_all_drugs(drugs, drug_mut_enets, drug_models, drug_loaders)
```
